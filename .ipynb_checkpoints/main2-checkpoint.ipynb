```python

#populate maps_info with more info for header
for map_id in maps_info:
    get_beatmaps_response = get_beatmaps(map_id)
    
    #---song's text info---
    #name
    maps_info[map_id]['name'] = get_beatmaps_response[0]['title']
    
    #artist
    maps_info[map_id]['artist'] = get_beatmaps_response[0]['artist']
    
    #creator
    maps_info[map_id]['creator'] = get_beatmaps_response[0]['creator']
   
    #difficulty name
    maps_info[map_id]['diff name'] = get_beatmaps_response[0]['version']
    
    #thumbnail
    map_url = mineThumbnail("https://old.ppy.sh/b/" + map_id)
    maps_info[map_id]['img'] = map_url
    
    #mod
    maps_info[map_id]['mod'] = mods_for_each_map[map_id]
    
    #---song stats---
    #length
    raw_length = int(get_beatmaps_response[0]['total_length'])
    mins = str(round(raw_length / 60))
    secs = raw_length % 60
    if secs < 10:
        secs = '0' + str(secs)
    else:
        secs = str(secs)
    length = mins + ':' + secs
    maps_info[map_id]['length'] = length
    
    #max_combo
    maps_info[map_id]['max_combo'] = int(get_beatmaps_response[0]['max_combo'])
    
    #bpm
    maps_info[map_id]['bpm'] = round(float(get_beatmaps_response[0]['bpm']), 0)
    
    #SR
    maps_info[map_id]['sr'] = round(float(get_beatmaps_response[0]['difficultyrating']), 2)
    
    #CS
    maps_info[map_id]['cs'] = float(get_beatmaps_response[0]['diff_size'])
    
    #HP
    maps_info[map_id]['hp'] = float(get_beatmaps_response[0]['diff_drain'])
    
    #OD
    maps_info[map_id]['od'] = float(get_beatmaps_response[0]['diff_overall'])
    
    #AR
    maps_info[map_id]['ar'] = float(get_beatmaps_response[0]['diff_approach'])
    
```

```python
def getNextEmptyRow(column, spreadsheetId, sheetName, offset):
    range_ = '%s!%s:%s' % (sheetName, column, column)
    request = service.spreadsheets().values().get(spreadsheetId=spreadsheetId, range=range_)
    response = request.execute()
    if 'values' not in response:
        return offset + 1
    #pprint(response)
    maps_col = response['values']
    empty_row_idx = len(maps_col) + offset
    return empty_row_idx
```
